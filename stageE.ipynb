```python
# Resample the data with daily frequency and sum the values for each day
daily_data = df_copy_ffill.resample('D').sum()
```

```python
daily_data.head()
```

```python
# Determine the length of the dataset
length = len(daily_data)

# Define the number of days for the test set (last 300 days)
test_days = 300

# Calculate the index to split the data
split_index = length - test_days

# Split the data into train and test sets
train_data = daily_data.iloc[:split_index]
test_data = daily_data.iloc[split_index:]
```
